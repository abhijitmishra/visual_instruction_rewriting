```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score
from nltk.translate.bleu_score import SmoothingFunction


nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')

test_predicted = pd.read_csv("all_test_predicted.tsv", sep='\t')

# Function to import TSV file
def import_tsv_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Sample TSV file (replace 'path_to_your_file.tsv' with the actual file path)
tsv_file = import_tsv_file('all_test_predicted.tsv')

# Split the TSV file into lines and ignore the header
lines = [line.split('\t') for line in tsv_file.strip().split('\n')[1:]]

# Initialize NLG evaluation metrics
bleu_scores = []
meteor_scores = []
rouge_scores = []

# Iterate through each line in the TSV file
for line in lines:
    rewritten_question = line[2]
    predicted = line[3]

    # Tokenize the reference and hypothesis sentences
    reference_tokens = word_tokenize(rewritten_question)
    predicted_tokens = word_tokenize(predicted)

    # Compute BLEU score
    reference = [reference_tokens]
    candidate = predicted_tokens
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=SmoothingFunction().method1)
    bleu_scores.append(bleu_score)

    # Compute METEOR score
    meteor_score_val = meteor_score([reference_tokens], predicted_tokens)
    meteor_scores.append(meteor_score_val)

    # Compute ROUGE score
    rouge = Rouge()
    rouge_score = rouge.get_scores(rewritten_question, predicted)[0]['rouge-l']['f']
    rouge_scores.append(rouge_score)

    # Compute BERTScore
    # scorer = BERTScorer(model_type='bert-base-uncased')
    # P, R, F1 = scorer.score([candidate], [reference])
    # print(P)
#     bert_scores.append(bert_score.item())

# Print computed scores
test_predicted["BLEU Scores"] = bleu_scores
test_predicted["METEOR Scores"] = meteor_scores
test_predicted["ROUGE Scores"] = rouge_scores
```
