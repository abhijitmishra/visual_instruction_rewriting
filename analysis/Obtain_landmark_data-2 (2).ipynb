We obtain landmarks URL for wikipedia pages from Google landmark dataaset https://github.com/cvdfoundation/google-landmark

We specifically take help of `train_label_to_hierarchical.csv` data to extract images and titles and caterogies. We restrict image extraction to man-made categories.


```python
import urllib.parse

import requests
from bs4 import BeautifulSoup

import requests
from bs4 import BeautifulSoup
import os
import urllib.parse
import pandas as pd 



df = pd.read_csv("train_label_to_hierarchical.CSV")
links_df = df['category']
links = list(links_df)[0:10]


# Function to download images from a Wikimedia Commons category page
def download_images_from_category_page(category_url, folder_path):
    try:
        # Make a request to get the HTML content of the category page
        response = requests.get(category_url)
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the image links on the page
        image_links = soup.find_all("div", class_="gallerytext")

        # Iterate over the image links
        for link in image_links:
            image_url = link.find("a")["href"]
            image_url = urllib.parse.urljoin("https://commons.wikimedia.org/", image_url)
            # Download the image
            download_image(image_url, folder_path)
    except Exception as e:
        print(f"Failed to download images: {str(e)}")

# Function to download an image
def download_image(image_url, folder_path):
    try:
        # Make a request to get the HTML content of the image page
        response = requests.get(image_url)
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the <a> tag containing the image URL
        image_anchor = soup.find("a", class_="internal")
        if image_anchor and 'href' in image_anchor.attrs:
            # Extract the image URL
            image_href = image_anchor['href']
            print(image_href)
            # Download the image
            image_name = os.path.basename(image_href)
            image_path = os.path.join(folder_path, image_name)
            with open(image_path, 'wb') as f:
                headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
                response = requests.get(image_href, headers=headers)
                f.write(response.content)
            print(f"Downloaded image: {image_name}")
        else:
            print("Image anchor or href attribute not found.")
    except Exception as e:
        print(f"Failed to download {image_url}: {str(e)}")




def extract_title(encoded_url):
    # Decode the URL string
    decoded_url = urllib.parse.unquote(encoded_url)

    # Extract the last element
    last_element = decoded_url.split("/")[-1]
    last_element = last_element.replace("Category:","")
    return last_element



for link in links:
    url = link
    title = extract_title(url)
    dest = f"C:/Users//Document/2024_spring/Captstone/scraping/training_images/{title}"
    if not os.path.exists(dest):
        os.makedirs(dest)
    download_images_from_category_page(url,dest)
```
